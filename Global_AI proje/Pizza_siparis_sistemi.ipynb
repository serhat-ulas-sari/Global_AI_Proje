```python
# Öncelikle Pizza süper sınıfını oluşturup açıklama ve fiyat fonksiyonlarını yazdım
class Pizza(object):
    def get_description(self) :
        self.klasik = " Sucuk,Mozarella peyniri ,zeytin ve sos"
        self.margarita = " Mozarella peynir ve sos"
        self.turk = " Lahmacun "
        self.sade ="Kaşar peynir, sucuk ve sos"
    def get_cost(self):
        self.f_klasik =20
        self.f_margarita = 15
        self.f_turk = 13
        self.f_sade =17
```

```python
# süper sınıftan "super()" metodunu kulanarak ihtiyacım olan fonksiyonları aldım
class klasikpizza(Pizza):
    def acıklamaK(self):
        super().get_description()
        super().get_cost()
        print("Klasik pizza içerik :",self.klasik)
        print("Klasik pizza fiyatı {} tl".format(self.f_klasik))
class margaritapizza(Pizza):
    def acıklamaM(self):
        super().get_description()
        super().get_cost()
        print("Margarita piza içerik : ", self.margarita)
        print("margarita pizza fiyatı {} tl".format(self.f_margarita))
class turkpizza(Pizza):
    def acıklamaT(self):
        super().get_description()
        super().get_cost()
        print("Türk piza içerik : ", self.turk)
        print("Mürk pizza fiyatı {} tl".format(self.f_turk))
        
class sadepizza(Pizza):
    def acıklamaS(self):
        super().get_description()
        super().get_cost()
        print("Sade piza içerik : ", self.sade)
        print("Sade pizza fiyatı {} tl".format(self.f_sade))
```

```python
# pizzalarda yaptığım gibi soslar için aynı mantıkla süper sınıf oluşturdum
class Sos(object):
    def get_description_sos(self):
        self.Zeytin = " Egenin en yaşlı ve güzel zeytin ağaçlarından toplanarak pizamıza değer katıyor"
        self.Mantarlar ="Düz kültür mantarı"
        self.KeciPeyniri ="Diyarbakırın ilk bahar aylarında taze otlaklşardan beslenilen keçilerimizin enfes peyniri"
        self.Et ="Tabi kuzu eti kulanırız kokusuna dayanamıyorsanız eklemeyin"
        self.Sogan ="Arpacık soğan"
        self.Misir="Konserve pek numarası yok"
    
    def get_cost_sos(self):
        self.f_Zeytin = 5
        self.f_Mantarlar =4
        self.f_KeciPeyniri =10
        self.f_Et =12       
        self.f_Sogan=2
        self.f_Misir=2
```

```python
class Zeytin (Sos ):
    def acıklamasosZ(self):
        super().get_description_sos()
        super().get_cost_sos()
        print("Zeytin:",self.Zeytin)
        print("Zeytinin Fiyatı {} tl".format(self.f_Zeytin))

class Mantarlar (Sos):
    def acıklamasosMa(self):
        super().get_description_sos()
        super().get_cost_sos()
        print("Mantarlar:",self. Mantarlar)
        print("Mantarlar Fiyatı {} tl".format(self.f_Mantarlar))

class KeciPeyniri (Sos):
    def acıklamasosK(self):
        super().get_description_sos()
        super().get_cost_sos()
        print("KeciPeyniri:",self.KeciPeyniri)
        print("KeciPeyniri Fiyatı {} tl".format(self.f_KeciPeyniri))
        
class Et (Sos):
    def acıklamasosE(self):
        super().get_description_sos()
        super().get_cost_sos()
        print("Et:",self.Et)
        print("Et Fiyatı {} tl".format(self.f_Et))
        
class Sogan(Sos):
    def acıklamasosS(self):
        super().get_description_sos()
        super().get_cost_sos()
        print("Sogan:",self.Sogan)
        print("Sogan Fiyatı {} tl".format(self.f_Sogan))

class Misir (Sos):
    def acıklamasosMi(self):
        super().get_description_sos()
        super().get_cost_sos()
        print("Misir:",self.Misir)
        print("Misir Fiyatı {} tl".format(self.f_Misir))
```
